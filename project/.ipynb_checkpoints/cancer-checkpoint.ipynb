#### 전세계 암의 분포를 설명

> 전세계 암의 분포 시각화: 데이터셋을 사용하여 전세계 각 국가에서 발생한 암의 종류 및 발생 빈도를 시각화할 수 있습니다. 이를 통해 전세계적으로 어떤 종류의 암이 가장 흔한지를 파악할 수 있습니다.

> 암의 발생 추이 분석: 연도별로 암의 발생 추이를 분석하여 특정 암의 발생 빈도가 시간에 따라 어떻게 변화하는지를 확인할 수 있습니다. 이를 통해 특정 암이 급증하거나 감소하는 추세를 발견할 수 있습니다.

> 지역별 암 발생 비교: 전세계의 다양한 국가들 간에 암의 발생률을 비교하여 특정 지역이 다른 지역보다 더 높은 암 발생률을 가지고 있는지를 확인할 수 있습니다. 이를 통해 암 발생에 영향을 미치는 지역적 요인을 파악할 수 있습니다.

#### 📊연도별, 대륙별 총 암발병 수 시각화

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# 경고를 무시
warnings.filterwarnings("ignore", message="use_inf_as_na option is deprecated")

# 대륙별로 암에 대한 사망자 수의 추이를 시각화
plt.figure(figsize=(8, 5))
sns.lineplot(data=year_continent_df, x='Year', y='Total', hue='Continent')
plt.title('Total Cancer Deaths by Continent over Time')
plt.xlabel('Year')
plt.ylabel('Number of Deaths')
plt.tight_layout()
plt.legend(title='Continent', loc='upper right')
plt.show()
```

##### 암 발병 수 높은 대륙
Asia > Europe > North America > Africa > South America > Oceania

각 대륙별 가장 높은 발병 수를 보이는 암 종류를 찾아보자.

```python
import warnings
# 경고 무시
warnings.filterwarnings("ignore", message="use_inf_as_na option is deprecated")

# 대륙별로 암에 대한 사망자 수의 추이를 시각화
plt.figure(figsize=(18, 12))

# 대륙 리스트
continents = year_continent_df['Continent'].unique()

# 1행에 3개씩 서브플롯 생성 및 시각화
num_rows = (len(continents) + 2) // 3
num_cols = min(len(continents), 3)

# 범례 표시를 위한 빈 리스트 생성
handles, labels = [], []

for i, continent in enumerate(continents):
    plt.subplot(num_rows, num_cols, i + 1)
    plt.title(f'Cancer in {continent} over Time')
    
    # 해당 대륙의 데이터 추출
    subset_df = year_continent_df[year_continent_df['Continent'] == continent]
    
    # 처음 다섯 개의 서브플롯에서는 범례를 표시하지 않음
    if i >= 5:
        # 각 암에 대한 시각화
        for cancer in cancer_columns:
            sns.lineplot(data=subset_df, x='Year', y=cancer, label=cancer)
            # 마지막 대륙에서만 범례 핸들과 레이블 추가
            if i == len(continents) - 1:
                line = plt.gca().lines[-1]
                handles.append(line)
                labels.append(cancer)
    else:
        # 각 암에 대한 시각화 (범례 없음)
        for cancer in cancer_columns:
            sns.lineplot(data=subset_df, x='Year', y=cancer)
    
    plt.xlabel('Year')
    plt.ylabel('Number of Patients')
    # plt.xticks(rotation=45)

# 맨 오른쪽 상단에 범례 추가
plt.legend(handles, labels, bbox_to_anchor=(1.05, 1), loc='upper left', title='Cancer Type')

plt.tight_layout()
plt.show()
```

```python
import warnings

# 경고를 무시
warnings.filterwarnings("ignore", message="use_inf_as_na option is deprecated")

# 대륙별로 암에 대한 사망자 수의 추이를 시각화
plt.figure(figsize=(18, 12))

# 대륙 리스트
continents = year_continent_df['Continent'].unique()

colors = ['#FF0000', '#008000', '#4B0082']

# 1행에 3개씩 서브플롯 생성 및 시각화
num_rows = (len(continents) + 2) // 3
num_cols = min(len(continents), 3)

for i, continent in enumerate(continents):
    plt.subplot(num_rows, num_cols, i + 1)
    plt.title(f'Top3 Cancer in {continent} over Time')
    
    # 해당 대륙의 데이터 추출
    subset_df = year_continent_df[year_continent_df['Continent'] == continent]
    
    # 대륙 내에서 암 발병 수가 높은 상위 3개의 암 추출
    top_cancers = subset_df.iloc[:, 2:-1].sum().nlargest(3).index.tolist()
    
    # 대륙 내에서 암 발병 수가 높은 상위 3개의 암에 대해 그래프 그리기
    for j, cancer in enumerate(top_cancers):
        sns.lineplot(data=subset_df, x='Year', y=cancer, label=None, color=colors[j])
        # 선 위에 암의 종류를 표시
        max_val = subset_df[cancer].max()
        max_year = subset_df.loc[subset_df[cancer] == max_val, 'Year'].iloc[0]
        plt.text(max_year, max_val, cancer, fontsize=9, ha='center', va='bottom', color=colors[j])
    
    plt.xlabel('Year')
    plt.ylabel('Number of Patients')

plt.tight_layout()
plt.show()
```

#### 🤔아프리카 대륙의 발병률 1위 암종류가 간암인 이유?

아프리카 대륙에서 간암 발생률이 높은 이유는 여러 가지 요인으로 설명될 수 있습니다  
 일반적으로 간암의 주요 원인은 간염이며, 이는 B형 및 C형 간염 바이러스로 인한 만성 간염과 관련이 있습니다  
 아프리카 대륙에서는 B형 및 C형 간염의 발병률이 높은 것으로 알려져 있으며, 이는 주로 감염 예방 접종의 접근성이 부족하고 감염된 혈액의 전파가 흔하기 때문일 수 있습니다  
발암 물질에 의해 인간 유전자에 남겨진 흔적의 한 예로 아프리카 사하라 지역과 동아시아의 간암 환자에서 TPS3 유전자의 codon249 변이를 들 수 있다.  
이들 지역의 간암은 간염 바이러스와 곰팡이의 한 종류인 aflatoxins 섭취에 의한 것이며, 동물/세포주 실험에서 aflatoxin은 codon249 변이를 유발하는 것이 밝혀졌다.  
이런 변이는 aflatoxin 노출이 낮은 지역의 간암에서는 발견되지 않는 지역적 특징이 있다.

#### 🤔전세계 발병률 1위 암종류가 폐암인 이유?
1. 흡연
> 흡연은 폐암의 가장 중요한 발병 요인입니다.  
> 담배에서는 7,000종 가량의 유해물질이 발견되는데, 이 가운데 발암물질로 알려진 것이 60여종 이상입니다.  
> 흡연자는 비흡연자에 비해 폐암에 걸릴 위험이 15~80배까지 증가합니다.  
> 흡연과 폐암의 상관관계는 담배를 피우는 양이 많을수록, 일찍 흡연을 시작할수록, 흡연 기간이 길수록 커집니다.  
> 또한 피우는 형태와도 관련이 있어서, 담배 연기를 들이마시는 깊이에 따라 위험도가 달라집니다.  

2. 간접 흡연
> 간접흡연은 비흡연자가 흡연자와 같이 생활하거나 그 주위에 있으면서 자신의 의지와 무관하게 담배 연기를 흡입하는 것으로, 직접흡연과 마찬가지로 폐암을 일으킬 수 있습니다.  
> 직접 담배를 피우지 않아도 간접흡연에 노출될 경우, 폐암에 걸릴 위험도는 1.2~2배까지 증가합니다.  
> 담배 연기는 두 종류로 나뉩니다.  
> 그 하나는 타고 있는 담배의 끝에서 바로 나오는 부류연(副流煙), 흔히 ‘생으로 태우는 연기’라고 하는 것이고, 또 하나는 흡연자가 들이켰다가 내뿜는 주류연(主流煙)입니다.  
> 잘 알려진 몇 종의 발암물질은 주류연보다 오히려 부류연에 훨씬 짙은 농도로 존재하는데, 간접흡연자는 대체로 주류연보다 부류연에 많이 노출되는 것으로 알려졌습니다(간접흡연에서 부류연의 비율이 85%라는 통계도 있습니다).  

3. 석면 등과 직업적 요인
> 석면과 연관된 폐암은 직업상 노출 때문인 경우가 많습니다.  
> 석면은 건축 자재(예전의 슬레이트 지붕 등), 저밀도 단열재, 전기 절연재, 방화재 등 다용도로 쓰여 왔는데, 석면에 지속적으로 노출되면 10~35년 정도의 잠복기를 거쳐 폐암이 발병하는 것으로 알려졌습니다.  
> 게다가 흡연자라면 폐암의 위험이 훨씬 더 커지기 쉽습니다.  

4. 직업적 요인
> 석면 이외에도 결정형 유리규산 분진에 노출되면 폐암 발생 위험이 증가합니다.  
> 비소, 베릴륨, 카드뮴, 6가크롬, 니켈 등의 중금속에 노출되는 경우 폐암 발생 위험이 증가합니다.  
> 콜타르 피치, 비스-클로로메틸에테르(bis-chloromethylether), 검댕 같은 화기물질에 노출되는 경우에도 폐암 발생 위험이 증가합니다. 
> 알루미늄 생산, 코크스(cokes) 생산, 주물업, 도장공과 같이 특정 작업에 종사하는 경우에도 폐암 위험이 증가합니다.  

5. 방사성물질
> 모든 종류의 방사성 동위원소는 발암 원인이 될 수 있습니다.  
> 우라늄은 소세포폐암의 발생과 밀접한 연관을 보이며, 특히 흡연자에게서 발생 빈도가 현저히 증가합니다.  
> 라돈은 라듐이 토양이나 암석, 물 속에서 붕괴할 때 발생하는 무색무취의 방사성 가스로, 대개 지표면을 통해 건물 내부 등으로 들어갑니다.  
> 흡연에 다음가는 폐암 발생 원인으로 추정됩니다.  
> 단순 X-선 촬영이나 전산화단층촬영(CT) 같은 방사선학적 검사에서 쐬는 방사선량은 미미하므로 폐암의 발생 원인이 되지 않습니다.  

6. 환경적 요인
> 디젤 연소물, 대기오염 먼지 중에는 다환방향족탄화수소(PAHs), 중금속 등 발암 물질이 함유되어 있으며, 장기간 다량의 노출이 있는 경우 폐암 발생 위험을 증가시킵니다.  
> 미세먼지는 WHO 세계보건기구에서 정한 1군 발암 물질로서, 미세 먼지가 높은 지역에서의 폐암 발생 위험이 증가한다고 보고되고 있습니다.  

7. 유전적 요인
> 폐암은 대부분 후천적인 유전자 이상 때문에 발생하며, 선천적인 유전자 이상에 의한 경우는 드물다고 알려졌습니다.  
> 가족력이 있는 경우엔, 다시 말해서 가족 중에 폐암 환자가 있으면 그렇지 않은 사람보다 발병 위험이 2~3배 정도 되는 것으로 보고되었습니다.

#### 🤔폐암에 흡연이 큰 영향을 미치는 이유?
- 흡연은 폐에 유해한 화학물질을 노출시키며, 이러한 화학물질은 폐 조직을 손상시키고 암을 유발할 수 있습니다.  
- 특히 담배 연기에 함유된 카르실렌(암을 유발하는 화합물)은 폐암의 주요 원인 중 하나로 알려져 있습니다.  - 
따라서 흡연은 폐암 발병 위험을 크게 증가시키는 요인 중 하나이며, 폐암 예방을 위해 흡연을 피하는 것이 매우 중요합니다..

#### 데이터 정보
> Cancer: 암 발병 여부

> Treatment_grps: 받은 1차 치료 "OLTx"(정위 간 이식), "절제", "절제", "TACE""(경동맥 화학색전술), "SIRT"(선택적 내부 방사선 요법), "의료", " 지지적 치료"  
> Survival_fromMDM: 다학제적 회의로부터의 생존  
> Alive_Dead : "살아있는", "죽은"  
> Type_of_incidental_finding: "1차 진료-일상 진료", "2차 진료-일상", "1차 진료-급성", "2차 진료-급성"  
> Surveillance_programme: 공식 감시 프로그램에 참여 중인 환자  
> Surveillance_effectiveness: 전년 대비 감시 준수("일관됨", "일관되지 않음", "누락됨")  
> Mode_of_surveillance_detection: 사고 감시 테스트 모드 "US"(초음파), "AFP 단독"(알파-태아단백질 단독), "CT/MRI"  
> Time_diagnosis_1st_Tx:  
> Date_incident_surveillance_scan: ("Y", "N")  
> PS: 성능 상태 0, 1, 2, 3, 4  
> Time_MDM_1st_treatment : 다학제 회의 1차 진료까지의 시간  
> Time_decisiontotreat_1st_treatment : 1차 치료 시기 결정  
> Prev_known_cirrhosis: "Y", "N"  
> Months_from_last_surveillance: 마지막 감시로부터 개월수

### 📊 간암 그래프 시각화
- 간암과 만성질환의 관련성? (아프리카 대륙과의 관련성?)

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 3, figsize=(20, 6))
hue_order = ['M', 'F']

# 히스토그램: bar 그래프 + 분포도(선)
# y축은 count이다.(분포를 나타내는 그래프이기 때문)
# multiple을 사용하지 않으면, 색이 섞여서 나타나는게 default
# KDE: 선 그래프(분포 확인), default=False이므로 True 옵션을 주어야함.
sns.histplot(ax=axes[0], x='Age', data=lc_df, hue='Gender', multiple='dodge', kde=True, hue_order=hue_order, stat='density')
sns.histplot(ax=axes[1], x='Age', data=lc_normal_df, hue='Gender', multiple='dodge', kde=True, hue_order=hue_order, stat='density')
sns.histplot(ax=axes[2], x='Age', data=lc_cancer_df, hue='Gender', multiple='dodge', kde=True, hue_order=hue_order, stat='density')

axes[0].set_title('All')
axes[1].set_title('Normal')
axes[2].set_title('Cancer')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 3, figsize=(20, 6))
hue_order = ['Y', 'N']

# 전체 환자, 기저질환 여부 count
sns.histplot(ax=axes[0], x='Age', data=lc_df, hue='Cirrhosis', multiple='dodge', kde=True, hue_order=hue_order)
# 암환자 일 때, 기저질환 여부 count
sns.histplot(ax=axes[1], x='Age', data=lc_cancer_df, hue='Cirrhosis', multiple='dodge', kde=True, hue_order=hue_order)
# 암환자가 아닐 때, 기저질환 여부
sns.histplot(ax=axes[2], x='Age', data=lc_normal_df, hue='Cirrhosis', multiple='dodge', kde=True, hue_order=hue_order)

axes[0].set_title('All')
axes[1].set_title('Cancer')
axes[2].set_title('Normal')

plt.show()
```

```python
# 간암이 발병한 년도에 대한 시게열 그래프
import pandas as pd
import matplotlib.pyplot as plt

# 'Month' 열을 년도와 월로 파싱하여 datetime 형식으로 변환하고, 새로운 열 생성
lc_cancer_df['Date'] = pd.to_datetime(lc_df['Month'], format='%b-%y', errors='coerce')

# 연도별 간암 발병 건수 집계
yearly_counts = lc_cancer_df['Date'].dt.to_period('M').value_counts().sort_index()

# 시계열 그래프 그리기
plt.figure(figsize=(10, 6))
yearly_counts.plot(kind='line', color='b')
plt.title('Liver Cancer Incidence Over the Years')
plt.xlabel('Year-Month')
plt.ylabel('Number of Cases')
plt.tight_layout()
plt.show()
```

#### 📍간암 데이터 분석
🤒간암 데이터 분석 결과는 다음과 같습니다.

1. 암을 겪고 있는 사람들과 기저질환의 관련성
   > 암을 겪고 있는 사람들 중 대부분이 간 기저질환이 있습니다.  
   > 이는 간암과 기저질환 간에 강한 연관성이 있음을 시사합니다.     
   > 반면 암을 겪고 있지 않은 사람들 중에서는 기저질환을 겪고 있는 사람들의 비율이 거의 없습니다.  
   > 이는 간암과 기저질환 간의 관련성이 강하게 나타남을 보여줍니다.

2. 나이와 기저질환, 간암 발병률의 관련성
   > 70세 중반까지는 대부분의 사람들이 기저질환을 앓고 있으며, 이와 관련하여 간암 발병률이 높아집니다.  
   > 이는 연령과 기저질환 간에 간암 발병률이 높은 상관관계를 보여줍니다. 또한, 이는 아프리카 대륙에서 간암 환자가 많은 이유 중 하나이기도 합니다.  
   > 그러나 이후 나이부터는 기저질환을 앓고 있지 않아도 간암 발병률이 높아지는 것으로 나타납니다.  
   > 이는 70세 후반부터는 기저질환과 간암의 상관관계가 줄어들고, 다른 요인들이 간암 발병에 영향을 미친다는 것을 시사합니다.

3. 시기별 간암 발병 환자 수 차이 (🤔왜 2020년도 1월에 간암 환자 수가 대폭 하락하였을까?)
   > 2019년 12월까지는 높은 수를 보이던 간암 환자의 수가 2020년 01월에 확 낮아진 것을 볼 수 있습니다.  
   > 이는 코로나19 유행 이후 의료이용 감소로 인한 진단 감소 영향으로 추정할 수 있습니다.  
   > 실제로 대한민국에서의 2020년 신규 암 진료 환자수는 2019년 대비 3.0% 감소하였으며,
   >  2017-19년 동월 평균 대비 2020년 진료실인원은 3, 4월에 입원 16.4%, 외래 16.5% 감소하였으며 6, 7월에 감소폭이 각각 3.0%까지 줄었다가 이후 13.8%, 14.1%까지 차차 증가하였습니다.
   
- 이러한 분석 결과는 간암 발병에 영향을 미치는 요인들을 이해하는 데 도움이 됩니다.
- 특히 연령과 기저질환 간의 관계를 고려하여 간암 예방 및 조기 진단을 위한 대책을 마련하는 데 유용할 것입니다.
