### RFM 분석

#### 고객 분석
> item_id : 상품코드  
> status: 주문 상태  
> created_at: 주문 날짜  
> sku: 재고 관리에 필요한 데이터  
> price: 가격  
> qty_ordered: 주문 수량  
> grand_total: 총 결제 가격  
> increment_id: 자동 증가 id  
> category_name_1: 카테고리명  
> sales_commision_code: 세일 코드  
> discount_amout: 할인 가격  
> payment_method: 결제 방법  
> Working Date: 완료 날짜  
> BI status: business intelligence(의사 결정 프로세스에 사용될 데이터를 수집하고 분석하는 프로세스)  
> Year: 주문 년도  
> Month: 주문 월  
> Customer Since: 가입 날짜  
> M-Y: 연-월  
> FY: 회계 연도  
> Customer ID: 고객 id  
> Unnamed : 잘못 늘어난 데이터

#### 이상치 검사

```python
import pandas as pd
from datetime import datetime
#  purchase_date(구매 일자) '오늘' 보다 미래인 경우
# 'purchase_date'를 datetime 형식으로 변환
e_df['purchase_date'] = pd.to_datetime(e_df['purchase_date'])
e_df['purchase_date'].max()
```

```python
# unit_price, total_price, discount가 0보다 작은 경우
display(e_df[e_df['unit_price'] < 0])
display(e_df[e_df['total_price'] < 0])
display(e_df[e_df['discount'] < 0])
```

- total_price가 0보다 작은 데이터와 discount가 0보다 작은 데이터 존재함.
- 이를 대체하기 위한 방법
  > discount를 양수로 바꿔주기  
  > unit_price * quantity - discount로 재계산하여 total_price에 다시 넣어주기  

```python
# discount가 0보다 작은 경우를 필터링하여 양수로 바꾸기
e_df.loc[e_df['discount'] < 0, 'discount'] *= -1
display(e_df[e_df['discount'] < 0])
```

```python
# unit_price * quantity - discount로 재계산하여 total_price에 다시 넣기
e_df['total_price'] = e_df['unit_price'] * e_df['quantity'] - e_df['discount']

display(e_df[e_df['total_price'] < 0])

# total_price가 음수인 경우를 찾아서 total_price를 0으로 대체
negative_total_price_indices = e_df[e_df['total_price'] < 0].index
e_df.loc[negative_total_price_indices, 'total_price'] = 0

display(e_df[e_df['total_price'] < 0])
```

```python
# 각 고객별로 'status'가 'canceled'인 행과 'complete'인 행의 수 계산
customer_status_counts = e_df.groupby('customer')['status'].value_counts().unstack(fill_value=0)

# 'canceled'인 행의 수가 'complete'인 행의 수보다 많은 경우 해당 행의 인덱스를 식별
to_drop = customer_status_counts[customer_status_counts['canceled'] > customer_status_counts['complete']].index

# 해당 행 삭제
e_df = e_df[~e_df['customer'].isin(to_drop)]

e_df = e_df.reset_index(drop=True)

e_df
```

```python
e_df['category'].value_counts()
```

```python
# '/N'이라는 데이터가 있는 행 삭제 (카테고리 별로 분석할 때의 정확도를 위해)
e_df = e_df[~(e_df['category'] == '\\N')]
```

```python
e_df['category'].value_counts()
```

#### RFM 계산 시, 유의할 점
- 현재 e_df에는 주문 완료된 데이터와 주문 취소된 데이터가 섞여있다.
- 따라서, R과 F를 계산할 때에는 주문완료된 데이터만을 사용하고,
- M을 계산할 때는 '주문완료 - 주문취소 = 최종 결제 금액'으로 두어야하기 때문에 주문완료된 데이터와 주문 취소된 데이터를 둘 다 사용한다. 

### Recency

```python
e_complete_df= e_df[e_df['status'] == 'complete']
e_canceled_df= e_df[e_df['status'] == 'canceled']
```

```python
display(e_complete_df)
display(e_canceled_df)
```

```python
e_complete_df = e_complete_df.reset_index(drop=True)
e_canceled_df = e_canceled_df.reset_index(drop=True)
```

```python
display(e_complete_df)
display(e_canceled_df)
```

```python
# Recency는 고객별로 구매한 내역이 가장 최신인 날짜를 이용하여 계산되어야함
# e_complete_df 이용
import pandas as pd
from datetime import datetime

# 주문이 완료된(complete) 상태인 가장 최근의 날짜 찾기 + 1일
current_date = e_complete_df['purchase_date'].max() + pd.Timedelta(days=1)
print(current_date)

# 'e_complete_df'에서 각 고객별 최근 구매일 계산
customer_recency = e_complete_df.groupby('customer')['purchase_date'].max().reset_index()
customer_recency.columns = ['customer', 'recent_purchase_date']

# 'e_df'에 Recency 추가
e_df = pd.merge(e_df, customer_recency, on='customer', how='left')
e_df['Recency'] = current_date - e_df['recent_purchase_date']

e_df
```

### Frequency

```python
# 'customer'로 그룹화하여 완료 횟수 계산하여 Frequency 열 추가
e_complete_df['Frequency'] = e_complete_df.groupby('customer')['customer'].transform('count')
# 'customer'로 그룹화하여 취소 횟수 계산하여 Frequency 열 추가
e_canceled_df['Frequency'] = e_canceled_df.groupby('customer')['customer'].transform('count')
```

```python
display(e_complete_df)
display(e_canceled_df)
```

```python
# customer별로 (완료 횟수 - 취소 횟수) 하여 e_df의 customer가 같을 경우 해당 행에 Frequency 추가

# 각 고객별로 완료 횟수와 취소 횟수 계산
complete_counts = e_complete_df['customer'].value_counts()
canceled_counts = e_canceled_df['customer'].value_counts()

# Frequency 계산: 완료 횟수 - 취소 횟수
frequencies = complete_counts.sub(canceled_counts, fill_value=0)

# e_df에 Frequency 열 추가
e_df['Frequency'] = e_df['customer'].map(frequencies)

e_df
```

```python
e_df.info()
```

```python
e_df = e_df.drop(['recent_purchase_date'], axis=1)
e_df['Frequency'] = e_df['Frequency'].astype(int)
```

```python
e_df.info()
```

### Monetary

```python
# customer별로 (왼료된 total_price - 취소된 total_price) 하여 e_df의 customer가 같을 경우 해당 행에 Monetary 추가

# 각 고객별로 완료된 total_price의 합계와 취소된 total_price의 합계 계산
complete_total_price = e_complete_df.groupby('customer')['total_price'].sum()
canceled_total_price = e_canceled_df.groupby('customer')['total_price'].sum()

# Monetary 계산: 완료된 total_price - 취소된 total_price
monetary = complete_total_price.sub(canceled_total_price, fill_value=0)

# e_df에 Monetary 열 추가
e_df['Monetary'] = e_df['customer'].map(monetary)

e_df
```

```python
# Monetary가 0보다 작은 데이터 이상치로 간주 -> 삭제
e_df = e_df[e_df['Monetary'] >= 0]
# Monetary int로 형변환
e_df['Monetary'] = e_df['Monetary'].astype(int)
```

```python
e_df = e_df.reset_index(drop=True)
e_df
```

```python
e_rfm_df = e_df[['Recency', 'Frequency', 'Monetary']]
e_rfm_df
```

### 정규화
값을 0~1 사이로 변환시켜 모든 컬럼의 데이터가 평등하게 한다.

```python
from sklearn.preprocessing import MinMaxScaler

normalization = MinMaxScaler()

# timedelta 값을 일수로 변환하여 'Recency' 열에 할당
e_df['Recency'] = (current_date - e_df['purchase_date']).dt.days
e_rfm_df = e_df[['Recency', 'Frequency', 'Monetary']]

# RFM 데이터를 정규화
rfm_normalization = normalization.fit_transform(e_rfm_df)

# 정규화된 데이터를 데이터프레임으로 변환
e_rfm_df = pd.DataFrame(rfm_normalization, columns=e_rfm_df.columns)
e_rfm_df
```

```python
e_rfm_df['Recency'].value_counts()
```

```python
# Recency는 최근 구매일 수록 점수가 높아야함으로 1에서 빼준다
e_rfm_df['Recency'] = 1 - e_rfm_df['Recency']
e_rfm_df
```

```python
e_df['TotalScore'] = e_rfm_df['Recency'] + e_rfm_df['Frequency'] + e_rfm_df['Monetary']
e_df
```

```python
e_df = e_df.drop(columns=['Recency', 'Frequency', 'Monetary'], axis=1)
e_df
```

### 📊데이터 시각화

#### 회원 등급 나누기
- 'Bronze', 'Silver', 'Gold', 'Diamond', 'VIP'

```python
import numpy as np
# 4분위를 사용하여 척도를 나눈다
l1, l2, l3, l4 = np.percentile(e_df['TotalScore'], [20, 45, 70, 90])
print(l1, l2, l3, l4)
```

```python
def get_level(x):
    if x <= l1:
        return 5
    if x> l1 and x <= l2:
        return 4
    if x > l2 and x <= l3:
        return 3
    if x > l3 and x <= l4:
        return 2
    return 1
```

```python
e_df['Level'] = e_df['TotalScore'].apply(get_level)
```

```python
e_df['Level'] = e_df['Level'].replace([5, 4, 3, 2, 1], ['Bronze', 'Silver', 'Gold', 'Diamond', 'VIP'])
e_df
```

```python
print(e_df['Level'].value_counts())
```

#### 📍주목할 점

```python
# 등급별 상품 판매 일시 그래프
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 등급 리스트
levels = ['Bronze', 'Silver', 'Gold', 'Diamond']

# 전체 기간의 모든 날짜 생성
all_dates = pd.date_range(start=e_df['purchase_date'].min(), end=e_df['purchase_date'].max())

# 모든 날짜를 포함하는 새로운 데이터 프레임 생성
all_dates_df = pd.DataFrame({'purchase_date': all_dates})

# 그래프 영역 설정
fig, axes = plt.subplots(3, 2, figsize=(20, 20))

# 각 등급에 대한 그래프 그리기
for i, level in enumerate(levels):
    row = i // 2
    col = i % 2
    # 결제 완료이고 해당 등급의 데이터 추출
    level_data = e_df[(e_df['Level'] == level) & (e_df['status'] == 'complete')]
    
    # 등급의 구매 일자별 수 계산
    level_count = level_data['purchase_date'].value_counts().sort_index()
    
    # 모든 날짜를 포함하는 데이터 프레임과 병합하여 결측값을 채움
    level_count = all_dates_df.merge(pd.DataFrame({'purchase_date': level_count.index, 'count': level_count.values}),
                                     how='left', on='purchase_date')['count'].fillna(0)
    
    # 라인 플롯 그리기
    sns.scatterplot(ax=axes[row, col], x=all_dates_df['purchase_date'], y=level_count)
    axes[row, col].set_title('Number of ' + level + ' Members per Purchase Date', fontsize=16)
    axes[row, col].set_xlabel('Purchase Date')
    axes[row, col].set_ylabel('Number of ' + level + ' Members')

# 결제 완료인 VIP 등급에 대한 그래프 그리기
vip_data = e_df[(e_df['Level'] == 'VIP') & (e_df['status'] == 'complete')]
vip_count = vip_data['purchase_date'].value_counts().sort_index()

# 모든 날짜를 포함하는 데이터 프레임과 병합하여 결측값을 채움
vip_count = all_dates_df.merge(pd.DataFrame({'purchase_date': vip_count.index, 'count': vip_count.values}),
                               how='left', on='purchase_date')['count'].fillna(0)

# 라인 플롯 그리기
sns.scatterplot(ax=axes[2, 0], x=all_dates_df['purchase_date'], y=vip_count)
axes[2, 0].set_title('Number of VIP Members per Purchase Date', fontsize=16)
axes[2, 0].set_xlabel('Purchase Date')
axes[2, 0].set_ylabel('Number of VIP Members')

# VIP 등급을 그리는 subplot의 빈 칸 숨기기
axes[2, 1].axis('off')

plt.tight_layout()
plt.show()
```

#### 📍주목할 점
- Bronze나 Silver 등급에 비해, Diamon와 VIP등급은 보다 최근의 구매율이 높은 것⬆️을 볼 수 있다.
- 하지만, Diamon와 VIP등급을 제외한 등급들은 2018년의 구매가 없는 것⤵️을 볼 수 있다.

```python
import matplotlib.pyplot as plt

# 카테고리 별 결제된 총 금액
payment_per_category = e_complete_df.groupby('category')['total_price'].sum()

# 카테고리 별 환불된 총 금액
refund_per_category = e_canceled_df.groupby('category')['total_price'].sum()

# 그래프 그리기
plt.figure(figsize=(12, 6))

# 카테고리 목록
categories = payment_per_category.index

# 막대 그래프를 옆으로 나란히 그리기 위해 위치 조절
bar_width = 0.35
index = range(len(categories))

# 결제된 총 금액 막대 그래프
plt.bar(index, payment_per_category, width=bar_width, color='lightblue', label='Payment')

# 환불된 총 금액 막대 그래프를 옆으로 이동하여 그리기
plt.bar([i + bar_width for i in index], refund_per_category, width=bar_width, color='lightcoral', label='Refund')

plt.xlabel('Category', fontsize=12)
plt.ylabel('Total Amount', fontsize=12)
plt.title('Total Payment and Refund by Category', fontsize=14)
plt.xticks([i + bar_width / 2 for i in index], categories, fontsize=7)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# 카테고리 별 결제된 총 금액
payment_per_category = e_complete_df.groupby('category')['total_price'].sum()

# 카테고리 별 환불된 총 금액
refund_per_category = e_canceled_df.groupby('category')['total_price'].sum()

# 총 매출 - 총 환불 금액을 계산하여 순 매출 얻기
net_sales_per_category = payment_per_category - refund_per_category

# 그래프 그리기
plt.figure(figsize=(12, 6))

# 카테고리 목록
categories = net_sales_per_category.index

# 막대 그래프 그리기
plt.bar(categories, net_sales_per_category, color='skyblue')

plt.xlabel('Category', fontsize=12)
plt.ylabel('Net Sales', fontsize=12)
plt.title('Net Sales by Category', fontsize=14)
plt.xticks(fontsize=7)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# 카테고리 별 결제된 총 금액
payment_per_category = e_complete_df.groupby('category')['total_price'].sum()

# 카테고리 별 환불된 총 금액
refund_per_category = e_canceled_df.groupby('category')['total_price'].sum()

# 총 매출 - 총 환불 금액을 계산하여 순 매출 얻기
net_sales_per_category = payment_per_category - refund_per_category

# 상위 5개 카테고리 선택
top_categories = net_sales_per_category.nlargest(5)

# 그래프 그리기
plt.figure(figsize=(5, 5))

# 카테고리 목록
categories = top_categories.index

# 순 매출에 해당하는 데이터
net_sales_values = top_categories.values

# 파이 그래프 그리기
plt.pie(net_sales_values, labels=categories, autopct='%1.1f%%', startangle=140)

plt.title('Net Sales by Top 5 Categories', fontsize=14)
plt.axis('equal')
plt.tight_layout()

plt.show()
```

```python
# 연도별 매출 시계열 그래프
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 주문 날짜 및 매출/환불 계산
revenue_per_day = e_complete_df.groupby('purchase_date')['total_price'].sum()
refund_per_day = e_canceled_df.groupby('purchase_date')['total_price'].sum()

# 주문 날짜가 연속되도록 모든 날짜를 포함하는 DataFrame 생성
all_dates = pd.date_range(start=e_df['purchase_date'].min(), end=e_df['purchase_date'].max())
revenue_per_day = revenue_per_day.reindex(all_dates, fill_value=0)
refund_per_day = refund_per_day.reindex(all_dates, fill_value=0)

# 순 매출 계산
net_revenue_per_day = revenue_per_day - refund_per_day

# 각 연도별 데이터 추출
revenue_2016 = net_revenue_per_day['2016']
revenue_2017 = net_revenue_per_day['2017']
revenue_2018 = net_revenue_per_day['2018']

# 2016년 시계열 그래프
plt.figure(figsize=(15, 6))
plt.plot(revenue_2016.index, revenue_2016.values)
plt.title('Net Revenue per Day in 2016', fontsize=13)
plt.xlabel('Purchase Date', fontsize=10)
plt.ylabel('Net Revenue', fontsize=10)
plt.show()

# 2017년 시계열 그래프
plt.figure(figsize=(15, 6))
plt.plot(revenue_2017.index, revenue_2017.values)
plt.title('Net Revenue per Day in 2017', fontsize=13)
plt.xlabel('Purchase Date', fontsize=10)
plt.ylabel('Net Revenue', fontsize=10)
plt.show()

# 2018년 시계열 그래프
plt.figure(figsize=(15, 6))
plt.plot(revenue_2018.index, revenue_2018.values)
plt.title('Net Revenue per Day in 2018', fontsize=13)
plt.xlabel('Purchase Date', fontsize=10)
plt.ylabel('Net Revenue', fontsize=10)
plt.show()
```

#### 📍주목할 점
- 2016년과 2017년의 11~12월에 다른 월들 보다 더 많은⬆️ 구매가 있었던 것을 볼 수 있다.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 2016년 11월부터 12월까지의 데이터 추출
orders_2016 = e_complete_df[(e_complete_df['purchase_date'] >= '2016-11-01') & (e_complete_df['purchase_date'] <= '2016-12-31')]
# 2017년 11월부터 12월까지의 데이터 추출
orders_2017 = e_complete_df[(e_complete_df['purchase_date'] >= '2017-11-01') & (e_complete_df['purchase_date'] <= '2017-12-31')]

# 각 카테고리별 구매량 계산
revenue_2016 = orders_2016.groupby('category')['total_price'].sum().nlargest(5)
revenue_2017 = orders_2017.groupby('category')['total_price'].sum().nlargest(5)

# 그래프 그리기
plt.figure(figsize=(20, 8))

# 2016년 데이터 subplot
plt.subplot(1, 2, 1)
sns.barplot(x=revenue_2016.index, y=revenue_2016.values, palette='Paired')
plt.title('Category Revenues from Nov to Dec 2016', fontsize=13)
plt.xlabel('Category', fontsize=10)
plt.ylabel('Revenue', fontsize=10)

# 2017년 데이터 subplot
plt.subplot(1, 2, 2)
sns.barplot(x=revenue_2017.index, y=revenue_2017.values, palette='Paired')
plt.title('Category Revenues from Nov to Dec 2017', fontsize=13)
plt.xlabel('Category', fontsize=10)
plt.ylabel('Revenue', fontsize=10)

plt.tight_layout()
plt.show()
```

#### 📍주목할 점
- 2016년과 2017년 11~12월에 매출이 높은 상품 카테고리가 비슷하고, 2016년에 비해 2017년에 더 매출이 높은 것을 볼 수 있다.

1. 회원 등급 별
   > Diamond, VIP회원은 최근까지도 꾸준한 구매를 하고 있는 것을 볼 수 있었지만, Bronze와 Silver회원은 최근의 구매이력이 없었습니다.  
   > 따라서, Bronze와 Silver 회원들을 타겟팅하여 재구매를 유도할 수 있는 프로모션을 제공해야 합니다.  
   > Bronze나 Silver 등급의 고객은 주로 👕 Men's Fashion 카테고리를 선호합니다.  
   > 그러나 Diamond이나 VIP 등급의 고객은 💻 Mobiles & Tablets 카테고리에 더 많은 관심을 보입니다.  
   > 따라서, Bronze/Silver 고객을 타켓팅한 Men's Fashion 상품과, Diamond/VIP 고객을 타겟팅한 Mobile & Tablets 상품을 통한 마켓팅을 진행하여 매출 상승을 기대할 수 있습니다.
2. 상품 카테고리 별
   > Mobiles & Tablets 카테고리가 가장 높은 매출을 기록하고 있으므로 해당 카테고리의 인기 상품을 중점적으로 홍보하고 마케팅을 진행해야 합니다.  
   > 또한, 다른 카테고리들에 대해서도 고객들의 관심을 끌 수 있는 새로운 프로모션을 고려해야 합니다.
3. 환불된 상품 별
   > 가장 많이 팔리는 카테고리는 👕 Men's Fashion이지만, 가장 많이 환불된 카테고리는 💻 Mobiles & Tablets으로 나타났습니다.  
   > 이러한 상반된 결과는 각 카테고리의 인기와 관련하여 고려해야 할 중요한 요소입니다.  
   > Mobiles & Tablets 카테고리의 높은 환불율은 제품에 대한 고객 만족도와 관련이 있을 수 있으며, 이를 개선하는 것이 중요합니다.  
   > Men's Fashion 카테고리의 높은 판매량은 해당 카테고리에 대한 고객 수요의 높음을 시사합니다.  
   > 따라서 이러한 인기 상품을 효과적으로 마케팅하여 매출을 더욱 증대시킬 수 있습니다.  다4
