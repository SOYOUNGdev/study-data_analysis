#### 데이터 분석
> 당뇨병(이진) : 환자가 당뇨병인지 여부를 나타냅니다(예인 경우 1, 아니요인 경우 0).  
> 알코올 농도(연속) : 환자의 알코올 소비 수준을 나타냅니다.  
> 심박수(연속) : 분당 심박수로 환자의 심박수를 나타냅니다.  
> 혈중 산소 농도(연속) : 혈액 내 산소 포화도를 반영합니다.  
> 체온(연속) : 환자의 체온을 섭씨 단위로 나타냅니다.  
> 체중(연속) : 환자의 체중을 킬로그램 단위로 나타냅니다.  
> MRI 지연(연속) : MRI 스캔의 지연 시간을 나타내며, 이는 의학적 상태 진행을 나타낼 수 있습니다.  
> 치매(Binary) : 치매 유무(0)를 나타내는 목표변수로 사용됩니다.   사용됩니다.

```python
import pandas as pd

path = './datasets/Cancer Deaths by Country and Type Dataset.csv'
cd_world_df = pd.read_csv(path, sep=',')
display(cd_world_df)
cd_world_df.info()
```

```python
# 각 암에 대한 사망자 수 추출
cancer_columns = ['Liver cancer', 'Kidney cancer', 'Larynx cancer', 'Breast cancer', 'Thyroid cancer', 
                  'Stomach cancer', 'Bladder cancer', 'Non-melanoma skin cancer', 'Lip and oral cavity cancer', 
                  'Brain and nervous system cancer', 'Tracheal, bronchus, and lung cancer', 
                  'Gallbladder and biliary tract cancer', 'Malignant skin melanoma', 'Leukemia', 
                  'Hodgkin lymphoma', 'Multiple myeloma', 'Other cancers']

country_cancer_deaths = cd_world_df.groupby('Country')[cancer_columns].sum().reset_index()
print(country_cancer_deaths)
```
