### 고객 분석
> Customer_id: 고객 id  
> Age: 고객 나이  
> Gender: 고객 성별  
> Revenue_Total: 고객별 총 매출  
> N_Purchases: 현재까지 구매 수  
> Purchase_DATE: 최근 구매 날짜 (dd.mm.yy)  
> Purchase_VALUE: 최근 구매 가격  
> Pay_Method: 결제 방식(0: 전자 지갑, 1:카드, 2:PayPal, 3:기타)  
> Time_Spent: 웹사이트에서 보낸 시간(초)  
> Browser: 접속한 브라우저(0: Chrome, 1:Safari, 2:Edge, 3:기타)  
> Newsletter: 구독 여부(0: 구독 않함, 1: 구독함)  
> Voucher: 바우처 사용 여부(0: 사용 않함, 1: 사용함)

#### 이상치 검사

```python
# Revenue_Total(총 구매 가격), N_Purchases(총 구매 수량), Purchase_VALUE(최근 구매 가격), Time_Spent(사이트에서 보낸 시간), Age(나이) 
# 0보다 작은 경우
display(os_df[os_df['Revenue_Total'] < 0])
display(os_df[os_df['N_Purchases'] < 0])
display(os_df[os_df['Purchase_VALUE'] < 0])
display(os_df[os_df['Time_Spent'] < 0])
display(os_df[os_df['Age'] < 0])
```

```python
import pandas as pd
from datetime import datetime
#  Purchase_DATE(최근 구매 일자) '오늘' 보다 미래인 경우
# 'Purchase_DATE'를 datetime 형식으로 변환
os_df['Purchase_DATE'] = pd.to_datetime(os_df['Purchase_DATE'])

# 오늘 날짜
current_date = pd.to_datetime(datetime.today().date())
print(current_date)

os_df[os_df['Purchase_DATE'] > current_date]
```

```python
import numpy as np
from scipy.stats import iqr

Q1 = os_df.describe().T.loc['Revenue_Total', '25%']
Q3 = os_df.describe().T.loc['Revenue_Total', '75%']

iqr_value = iqr(os_df.Revenue_Total)

lower_bound = Q1 - 1.5 * iqr_value
upper_bound = Q3 + 1.5 * iqr_value

print(f'총 구매 금액 정상치 범위: {lower_bound} ~ {upper_bound}')

if lower_bound < 0:
    lower_bound = 0

print(f'수정된 정상치 범위: {lower_bound} ~ {upper_bound}')

# 정상치 범위 밖의 데이터가 존재하는지 확인 -> 결과: 모두 정상치 범위 내에 존재함
os_df[os_df['Revenue_Total'] > upper_bound]
```

#### 정규화
값의 범위를 0~1로 변환하여 모든 컬럼의 데이터가 평등하게 만든다.

```python
from sklearn.preprocessing import MinMaxScaler

normalization = MinMaxScaler()

# timedelta 값을 일수로 변환하여 'Recency' 열에 할당
os_df['Recency'] = (current_date - os_df['Purchase_DATE']).dt.days
os_rfm_df = os_df[['Recency', 'Frequency', 'Monetary']]

# RFM 데이터를 정규화
rfm_normalization = normalization.fit_transform(os_rfm_df)

# 정규화된 데이터를 데이터프레임으로 변환
os_rfm_df = pd.DataFrame(rfm_normalization, columns=os_rfm_df.columns)
os_rfm_df
```

```python
# Recency는 최근 구매일 수록 점수가 높아야함으로 1에서 빼준다
os_rfm_df['Recency'] = 1 - os_rfm_df['Recency']
os_rfm_df
```

```python
# 각 R,F,M합하여 TotalScore라는 새로운 feature 생성
os_df['TotalScore'] = os_rfm_df['Recency'] + os_rfm_df['Frequency'] + os_rfm_df['Monetary']
os_df
```

```python
# RFM score로 TotalScore 계산하였으므로 기존의 Recency, Frequency, Monetary 삭제한다.
os_df = os_df.drop(columns=['Recency', 'Frequency', 'Monetary'], axis=1)
os_df
```

#### 회원 등급 나누기
- Bronze, Silver, Gold, Diamond, VIP

```python
import numpy as np
# 4분위를 사용하여 척도를 나눈다
l1, l2, l3, l4 = np.percentile(os_df['TotalScore'], [20, 40, 70, 90])
print(l1, l2, l3, l4)
```

```python
def get_level(x):
    if x <= l1:
        return 5
    if x> l1 and x <= l2:
        return 4
    if x > l2 and x <= l3:
        return 3
    if x > l3 and x <= l4:
        return 2
    return 1
```

```python
os_df['Level'] = os_df['TotalScore'].apply(get_level)
```

```python
os_df['Level'] = os_df['Level'].replace([5, 4, 3, 2, 1], ['Bronze', 'Silver', 'Gold', 'Diamond', 'VIP'])
os_df
```

```python
print(os_df['Level'].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

order_columns = ['Bronze', 'Silver', 'Gold', 'Diamond', 'VIP']

sns.countplot(x='Level', data=os_df, palette='muted', order=order_columns)
```
